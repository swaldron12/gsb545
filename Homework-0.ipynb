```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Assumptions

# Estimated tax rate
tax_rate = 0.27

# Savings goal
savings_rate = 0.15

# Lifestyle expenses buffer
lifestyle_rate = 0.12

city_costs = {
    'San Francisco': {
        'Housing': 2700,
        'Food': 750,
        'Transportation': 220,
        'Utilities': 240,
        'Healthcare': 280,
        'Miscellaneous': 500
    },
    'Los Angeles': {
        'Housing': 2500,
        'Food': 650,
        'Transportation': 300,
        'Utilities': 220,
        'Healthcare': 260,
        'Miscellaneous': 450
    },
    'San Diego': {
        'Housing': 2300,
        'Food': 620,
        'Transportation': 240,
        'Utilities': 210,
        'Healthcare': 250,
        'Miscellaneous': 425
    }
}

assumptions_summary = pd.DataFrame({
    'Assumption': ['Effective tax rate', 'Savings rate', 'Lifestyle Rate'],
    'Value': [tax_rate, savings_rate, lifestyle_rate]
})

assumptions_summary
```

```python
# Estimate monthly spending by category for each city

monthly_cost_df = pd.DataFrame(city_costs).T

# Estimate total monthly and annual costs
monthly_cost_df['Total Monthly Cost'] = monthly_cost_df.sum(axis=1)
monthly_cost_df['Total Annual Cost'] = monthly_cost_df['Total Monthly Cost'] * 12

monthly_cost_df
```

```python
# Salary Calculation
salary_summary = monthly_cost_df[['Total Monthly Cost', 'Total Annual Cost']].copy()

salary_summary['Minimum Annual Salary Needed'] = (
    salary_summary['Total Annual Cost'] / (1 - tax_rate)
)

salary_summary['Desired Annual Salary'] = (
    salary_summary['Total Annual Cost'] * (1 + lifestyle_rate) / (1 - tax_rate - savings_rate)
)

salary_summary = salary_summary.round(2)
salary_summary
```

```python
# Plot minimum salary versus desired salary for each city

plot_df = salary_summary[['Minimum Annual Salary Needed', 'Desired Annual Salary']]

ax = plot_df.plot(
    kind='bar',
    figsize=(10, 6),
    color=['#4C78A8', '#F58518']
)
ax.set_title('Estimated Salary Needed After Completing an MSBA Program')
ax.set_xlabel('City')
ax.set_ylabel('Annual Salary (USD)')
ax.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()

```
